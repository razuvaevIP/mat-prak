1: Смоделировать выборку объема 1000 из дискретного распределения на множестве цифр 0, 1, 2, ..., 9 с весами 0.12, 0.3, 0.167, 0.24, 0.31, 0.54, 0.111, 0.02, 0.001, 0.2. По выборке построить гистограмму. Оптимизируйте алгоритм, упорядочив веса. Сравните время генерации выборки с неупорядоченными и упорядоченными весами.

```python
import matplotlib.pyplot as plt
import numpy as np
import time
import scipy.stats
from math import log
from scipy.stats import uniform
from scipy.stats import cauchy

def rng(m=2**32, a=1103515245, c=12345):
    rng.current = (a * rng.current + c) % m
    return rng.current / m
rng.current = 1

random = [rng() for i in range(1000)]
Numbers = np.arange(0,10)
NumbersSort = np.arange(0,10)
b1 = np.zeros(10)
b2 = np.zeros(10)
massiv_vesov = np.array([0.12, 0.3, 0.167, 0.24, 0.31, 0.54, 0.111, 0.02, 0.001, 0.2])
massiv_vesov = massiv_vesov/massiv_vesov.sum() 
```

```python
NumbersSort = massiv_vesov.argsort()[::-1]
massiv_vesovSort = np.sort(massiv_vesov)[::-1]
print(massiv_vesovSort)
print(NumbersSort)
```

```python
for i in range (1, 10):
    massiv_vesov[i] = massiv_vesov[i] + massiv_vesov[i-1]

for i in range (1, 10):
    massiv_vesovSort[i] = massiv_vesovSort[i] + massiv_vesovSort[i-1]
```

Неупорядочные веса:

```python
t1 = time.time()
for i in range(1000):
    for j in range (10):
        if ((j == 0 and random[i] < massiv_vesov[j]) or (j !=0 and random[i] < massiv_vesov[j] and random[i] >= massiv_vesov[j-1])):
            b1[j] = b1[j] + 1
            break

t2 = time.time()
print(t2 - t1) 
print(b1)
```

Упорядочные веса:

```python
t1 = time.time()
for i in range(1000):
    for j in range (10):
        if ((j == 0 and random[i] < massiv_vesovSort[j]) or (j !=0 and random[i] < massiv_vesovSort[j] and random[i] >= massiv_vesovSort[j-1])):
            b2[j] = b2[j] + 1
            break

t2 = time.time()
print(t2 - t1) 
print(b2)    
```

Алгоритм с упорядоченными весами работает быстрее.

6: Напишите функцию, которая моделирует случайное симметричное блуждание на двумерной решетке длины n c началом и концом в точке (0, 0). Приведите графики выборочных траекторий для n=100.

Получаем, что Box-Muller algorithm формирует выборку из независимых N(0,1) случаных величин.

```python
def Random_walk():
    x = np.array([(rng()-0.5)*2 for i in range(500)]) 
    y = np.where(x > 0)
    y = y[0][0:50]
    z = np.where(x < 0)
    z = z[0][0:50]

    A = np.zeros(120)
    A[y] += 1

    B = np.zeros(120)
    B[z] += 1

    W_ = A[(A != 0) | (B != 0)]
    Q_ = B[(A != 0) | (B != 0)]

    return np.cumsum(W_ - Q_)

x_1 = Random_walk()
x_2 = Random_walk()
y_1 = Random_walk()
y_2 = Random_walk()
x_1 = (x_1 + x_2) / 2 
x_2 = (x_1 - x_2) / 2
y_1 = (y_1 + y_2) / 2 
y_2 = (y_1 - y_2) / 2
fig, sub = plt.subplots(1, 2, figsize=(15, 5))
sub[0].plot(x_1, x_2, color = 'b')
sub[1].plot(y_1, y_2, color = 'r')
sub[0].axis('equal')
sub[1].axis('equal')
grid = sub[0].grid(True)
grid = sub[1].grid(True)
plt.show()
```
