```python
#task1:Plot the populations for the years in the period. Add legend and axis labels to the plot.

import numpy as np
import matplotlib.pyplot as plt
fname='http://scipy-lectures.org/_downloads/populations.txt'
a = np.loadtxt(fname, dtype=int, skiprows=1)
print(a)

X=np.linspace(1900,1920,21)
plt.plot(X,a[:,1],label='Hare',c='b')
plt.plot(X,a[:,2],label='Lynx',c='r')
plt.xlabel('Years')
plt.ylabel('Population')
plt.xticks(a[:,0][::2])
plt.legend()
plt.show()
```

Task 2: Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
X=np.linspace(1900,1920,21)
plt.bar(X,a[:,1]/(a[:,1]+a[:,2]),label='Hare')
plt.bar(X,a[:,2]/(a[:,1]+a[:,2]),label='Lynx',bottom=a[:,1]/(a[:,1]+a[:,2]))
plt.xlabel('Years')
plt.ylabel('Population')
plt.xticks(a[:,0][::2])
plt.legend()
plt.show()
```

Task 3:Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
fig, ax = plt.subplots(1, 3, figsize=(20,5))

X=np.linspace(1900,1920,21)
b=np.mean(a[:,1])
c=np.std(a[:,1])
ax[0].hist(a[:,1],label='Hare')
ax[0].axvline(x=b,label='mean value',c='r')
ax[0].legend()
print('Std of Hares:')
print (c)


n=np.mean(a[:,2])
m=np.std(a[:,2])
ax[1].hist(a[:,2],bins=8,label='Lynx')
ax[1].axvline(x=n,label='mean value',c='r')
ax[1].legend()
print('Std of Lynxs:')
print (m)


j=np.mean(a[:,3])
k=np.std(a[:,3])
ax[2].hist(a[:,3],bins=8,label='Carrot')
ax[2].axvline(x=j,label='mean value',c='r')
ax[2].legend()
print('Std of Carrots:')
print (k)
plt.show()
```

Task 4: Find which year each species had the largest population. Mark these years in the plot of populations.

```python
X=np.linspace(1900,1920,21)
m = np.argmax(a[:,1:], axis=0)

plt.scatter(a[m[0],0], a[m[0], 1], 150,c='r', marker = '.')
plt.plot(X, a[:,1], '-', label='Hare')

plt.scatter(a[m[1],0], a[m[1], 2], 150, c='r', marker = '.')
plt.plot(X, a[:,2], '-', label='Lynx')

plt.scatter(a[m[2],0], a[m[2], 3], 150, c='r', marker = '.')
plt.plot(X, a[:,3], '-', label='Carrot')

plt.xticks(a[:,0][::2])
plt.legend()
plt.show
```

Task 5:Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing ratio of "H" and "L" values obtained.

```python
arr = np.chararray(21)

mask = a[:,1] >= a[:,2]
arr[a[:,1] >= a[:,2]] = "H"
arr[a[:,1] < a[:,2]] = "L"
print(arr)

haresNumber = np.sum(np.ones(21)[mask])
lynxesNumber = arr.shape[0] - haresNumber

sizes = np.array([haresNumber, lynxesNumber])
plt.pie(sizes, labels = ["Hares", "Lynxes"])
plt.axis('equal')

plt.show()
```

Task 8: Plot the change in hare population and the number of lynxes. Find the correlation coefficient.

```python
plt.bar(a[1:,0], a[1:,1] - a[:-1,1], label="Change in hare population" )
plt.xticks(a[:,0][::2])
plt.legend()
plt.show()

plt.bar(a[:,0], a[:,2], label="Number of lynxes")
plt.xticks(a[:,0][::2])
plt.legend()
plt.show()

print("Correlation coefficient:")
print(np.corrcoef(a[:,1],a[:,2])[0,1])
```

```python
from sklearn.datasets import fetch_openml
mnist_data = fetch_openml('mnist_784')
```

```python
from sklearn.decomposition import PCA
```

```python
a = []

for i in np.arange(50, 401, 50):
    pca = PCA(n_components=i)
    X_transformed = pca.fit_transform(mnist_data.data.astype(float)) 
    a.append(pca.explained_variance_ratio_.sum())
```

```python
plt.plot(np.arange(50, 401, 50), a)
```
